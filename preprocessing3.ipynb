```python
import numpy as np
import pandas as pd
```

### Extracting features of 2018 movies from Wikipedia

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
df1 = pd.read_html(link , header=0)[2]
df2 = pd.read_html(link , header=0)[3]
df3 = pd.read_html(link , header=0)[4]
df4 = pd.read_html(link , header=0)[5]
```

```python
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
df['Opening'] = df['Opening'].str.replace(' ' , '')
```

```python
df
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = 'fce0af3409e6113c9b3c75aaf49341bb'
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genres(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id , tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " "
        for i in range (0 , len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
    else:
        np.NaN
```

```python
df['genres'] = df['Title'].map(lambda x: get_genres(str(x)))
```

```python
df
```

```python
df_2018 = df[['Title' , 'Cast and crew' , 'genres']]
```

```python
df_2018
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2018['actor_1_name'] = df['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2018['actor_2_name'] = df_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2018['actor_3_name'] = df_2018['Cast and crew'].map(lambda x: get_actor3(x))
```

```python
df_2018
```

```python
df_2018 = df_2018.rename(columns={'Title':'movie_title'})
```

```python
newdf_2018 = df_2018.loc[: , ['director_name' , 'actor_1_name' , 'actor_2_name' , 'actor_3_name' , 'genres' , 'movie_title']]
```

```python
newdf_2018
```

```python
newdf_2018.drop(1 , inplace=True)
newdf_2018.reset_index(drop=True , inplace=True)
```

```python
newdf_2018
```

```python
newdf_2018['actor_2_name'] = newdf_2018['actor_2_name'].replace(np.nan , 'unknown')
newdf_2018['actor_3_name'] = newdf_2018['actor_3_name'].replace(np.nan , 'unknown')
```

```python
newdf_2018['movie_title'] = newdf_2018['movie_title'].str.lower()
```

```python
newdf_2018['comb'] = newdf_2018['actor_1_name'] + ' ' + newdf_2018['actor_2_name'] + ' ' + newdf_2018['actor_3_name'] + ' ' + newdf_2018['director_name'] + ' ' + newdf_2018['genres']
```
